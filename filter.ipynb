```python
from utils import *
```

```python
img = readImage("./input/test.png")
showImage(img, figsize=(15, 15))
```

```python
kernel_lpls = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
], np.float32)
kernel_lpls
```

```python
dest = cv.filter2D(img, cv.CV_8UC3, kernel_lpls)
showImage(dest, figsize=(15, 15))
```

```python
size = 5
kernel_linear = np.ones([size, size], np.float32)/(size*size)
kernel_linear
```

```python
dest = cv.filter2D(img, cv.CV_8UC3, kernel_linear)
showImage(dest, figsize=(15, 15))
```

```python
dest = cv.blur(img, [5, 5], cv.CV_8UC3)
showImage(dest, figsize=(15, 15))
```

```python
dest = cv.GaussianBlur(img, [5, 5], -1)
showImage(dest, figsize=(15, 15))
```

```python
dest = cv.medianBlur(img, 5)
showImage(dest, figsize=(15, 15))
```

```python
dest = cv.bilateralFilter(img, 9, 75, 75)
showImage(dest, figsize=(15, 15))
```
