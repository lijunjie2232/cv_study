```python
img  = readImage('./input/test.png').astype(np.float64)
showImage(img)
```

```python
dest_x = cv.Scharr(img, cv.CV_64F, 1, 0)
dest_y = cv.Scharr(img, cv.CV_64F, 0, 1)
dest = cv.addWeighted(dest_x, 0.5, dest_y, 0.5, 0)
showImage(dest_x)
showImage(dest_y)
showImage(dest)
```

```python
dest_x = cv.Sobel(img, cv.CV_64F, 1, 0, ksize=3)
dest_y = cv.Sobel(img, cv.CV_64F, 0, 1, ksize=3)
dest = cv.addWeighted(dest_x, 0.5, dest_y, 0.5, 0)
showImage(dest_x)
showImage(dest_y)
showImage(dest)
```

```python
dest = cv.Laplacian(img, cv.CV_64F, ksize=5)
showImage(dest)
```

```python
img_gray = cv.cvtColor(img.astype(np.uint8), cv.COLOR_RGB2GRAY)
dest = cv.Canny(img_gray, 100, 200)
showImage_cv("", dest)
```
