```python
from utils import *
```

```python
img = readImage("../input/base.png")
showImage(img)
```

```python
colors = ('r', 'g', 'b')
for i, c in enumerate(colors):
    image = np.zeros(img.shape, dtype=np.uint8)
    image[:,:,i] = img[:,:,i]
    showImage(image, title=c)
    showImage(img[:,:,i], cmap='gray', title=c)
```

```python
plt.figure()
for i, c in enumerate(colors):
    hist = cv.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color=c, label=c)
plt.legend()
plt.show()
```

```python
from CaptureUtils import CaptureUtils
```

```python
cap = CaptureUtils(
    img = cv.cvtColor(img,cv.COLOR_RGB2BGR),
    mode = 1
)
```

```python
boxes = cap()
boxes
```

```python
box = boxes[0]
mask = np.zeros(img.shape, dtype=np.uint8)

mask[box[0][1]:box[1][1], box[0][0]:box[1][0]] = (255, 255, 255)
showImage(mask)
```

```python
showImage(cv.bitwise_and(img, mask))
```

```python
plt.figure()
for i, c in enumerate(colors):
    hist = cv.calcHist([img], [i], mask[:,:,i], [256], [0, 256])
    plt.plot(hist, color=c, label=c)
plt.legend()
plt.show()
```
