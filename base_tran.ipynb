```python
img = readImage('./input/perspect_test.png')
showImage(img)
```

```python
c = CaptureUtils(img, 0)
points = c.startCapture('q')
```

```python
points = np.array(points).astype(np.float32)
points
# [(119, 193), (580, 145), (826, 814), (426, 1040)]
```

```python
rows, cols = img.shape[:2]
points_after = np.array([[0,0], [cols, 0], [cols, rows], [0, rows]]).astype(np.float32)
points_after
```

```python
M = cv.getPerspectiveTransform(points, points_after)
M
```

```python
dest = cv.warpPerspective(img, M, img.shape[:2][::-1])
showImage(img)
showImage(dest)
```
